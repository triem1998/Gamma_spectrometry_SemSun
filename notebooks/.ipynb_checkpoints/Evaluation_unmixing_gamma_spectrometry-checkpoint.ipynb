## Read data

```python

def read_spectrum(fname):
    spec = []
    for line in open(fname, 'r'):
        spec.append(float(line.rstrip()))
    return np.array(spec)


def GetListData(name=0):
    PATH_DATA=os.path.join('../data/Simulation_steel_sphere', RN_NAME[name])
    list_thickness=[]
    list_data=[]

    for file in os.listdir(PATH_DATA):
        if file.endswith(".dat"):
            tmp_file=os.path.join(PATH_DATA, file)
            list_data+=[tmp_file]
    ## list thickness
    ## replace __ = . , (2__0=2.0mm)
    for tmp_file in list_data:
        pos=tmp_file.find("STEEL")
        tmp_thick=tmp_file.replace(tmp_file[:pos+5],"")
        tmp=tmp_thick.replace("mm.dat","")
        if "__" in tmp:
            tmp=tmp.replace("__",".")
        list_thickness+=[float(tmp)]
    ##sort data 
    ind=np.argsort(list_thickness)
    list_thickness=np.sort(list_thickness)
    list_data = [list_data[i] for i in ind]

    return list_data,list_thickness


def GetSpectra(name=0):
    list_data,list_thickness=GetListData(name)
    spec=[]
    for i in range(len(list_data)):
        spec += [read_spectrum(list_data[i])]
    spec=np.array(spec).T
    # cut 18kev 
    # 1024 channels, 1 channel=2kev
    spec = spec[18:2048+18,:]
    spec=spec[0::2,:]+spec[1::2,:]
    ## min value = 1e-6 of max value(avoid 0)   
    val_min=1e-6
    ind=spec<val_min*np.max(spec,axis=0).reshape(1,np.shape(spec)[1])
    spec=(1-ind)*spec+ind*np.ones(np.shape(spec))*val_min*np.max(spec,axis=0)
    return spec.T,list_thickness
```

## Define train test datasets

```python
def get_train_test(name,min_channel=0,max_channel=800,norm='1'):
    spec,list_thickness=GetSpectra(name)
    spec = spec.reshape(96,-1,1)
    np.random.seed(100)
    I1=np.random.permutation(24)+1
    I2=np.random.permutation(70)+25
    ## Training: 18 random of 24 low thicknesses, 51 random of 70 high thicknesses

    Itrain=np.concatenate((I1[:18],I2[:51]))
    Itrain=Itrain[np.random.permutation(len(Itrain))]
    Itrain=np.concatenate((np.array([0,95]),Itrain))
    Itest=np.concatenate((I1[18:],I2[51:]))

    Xtrain = normalize4(spec[Itrain,min_channel:max_channel,:],opt=norm)
    Xtest = normalize4(spec[Itest,min_channel:max_channel,:],opt=norm)
    AnchorPoints = Xtrain[[0,1]] ## anchor points: position 0 and 95 (1e-6 and 30 mm thickness) 
    return Xtrain,Xtest,AnchorPoints, Itrain,Itest,list_thickness
```

## Define spectral signature X using thickness

```python
def init_X(XMVP,position,list_radio,norm='1'):
    # position: position of thickness used to create X
    # list_radio: list of radionulcides in X
    list_signature=[]
    for i in range(len(list_radio)):
        spec,_=GetSpectra(name=list_radio[i])
        ind=spec<1e-6*np.max(spec,axis=1).reshape(np.shape(spec)[0],1)
        spec=(1-ind)*spec+ind*np.ones(np.shape(spec))*1e-6*np.max(spec,axis=1).reshape((96,1))
        spec=spec[:,:800]
        spec=normalize(spec,opt=norm)
        list_signature+=[spec[position[i]:position[i]+1,:]]
    X=np.concatenate(list_signature,axis=0).T
    X=(np.concatenate([XMVP,X],axis=1))
    return X
```
